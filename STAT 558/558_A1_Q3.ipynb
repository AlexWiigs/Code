The time to repair an electronic intrsument is a normally distributed random variable measured in hours. The repair time (in hours) for 16 such intruments chosen at random are as follows:

```r
hours <- c(159, 224, 222, 149, 280, 379, 362, 260, 101, 179, 168, 485, 212, 264, 250, 170)
n <- length(hours)
plot(density(hours))
```

**a)** You wish to know if the mean repair time exceeds 225 hours. Set up appropriate hypotheses for investigating this issue.

An approriate null and alternative hypthesis for the statement above can be stated as  
$$
H_0: \mu_0 = 225 \\ H_1: \mu > 225 
$$
Since we don't know the population standard deviation, we will estimate it. This means our test statistic will be 
$$
t = \frac{\bar y - 225}{\sqrt{S^2} / n} 
$$
where $S^2 = \sum_{i=1}^n\frac{x_i - \bar x}{n-1}$ is the estimated sample variance.

**b)** Using R and assuming $\alpha = 0.05$ test the hypotheses you formulated in part (a). What are your conclusions?  

With an $\alpha = 0.05$, we set will reject the null hypothesis in favour of alternative if $t_0 \geq t_{\alpha, n-1}$. 
$$
t_{\alpha, n-1} = t_{0.05, 15} = 1.753
$$

```r
qt(p=0.05, df = 15, lower.tail = FALSE)
```

```r
ybar <- mean(hours)
W <- 0
for (i in 1:16) { W <- W + (hours[i] - mean(hours) )^2}
S2 <- W / (n-1)
ybar
S2
```
