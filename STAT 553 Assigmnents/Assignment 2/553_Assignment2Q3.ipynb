```r
mu <- matrix(c(1,2))
Sigma <- matrix(c(3, -2, -2, 9), nrow = 2)

library(MASS)
set.seed(1992)
n <- 300
X <- mvrnorm(n = n, mu = mu, Sigma = Sigma)
head(X)
```

## a)

```r
plot(X)
```

## b)

```r
MLE_mu <- matrix(c((1/n) * sum(X[,1]), (1/n) * sum(X[,2])))
MLE_mu
```

```r
y_yprime <- matrix(c(0,0,0,0), ncol = 2)
for (i in 1:n) { 
    y_yprime <- y_yprime + X[i,] %*% t(X[i,])
}

MLE_S <- (1/n) * (y_yprime - n * MLE_mu %*% t(MLE_mu))
MLE_S
```

Double Check: (We should expect our function to have slightly smaller values because MLE divides by $n$ instead of $n-1$.)

```r
cov(X)
```

## C)

```r
Delta_Sqr <- rep(0,n)

for (i in 1:n) {
   Delta_Sqr[i] <- t((X[i,] - MLE_mu)) %*% solve(MLE_S) %*% (X[i,] - MLE_mu) 
}
Delta_Sqr
```

Double Check:

```r
mahalanobis(X, colMeans(X), cov(X))
```

Values are very similar.

Plotting the values of $\Delta^2$, we have:

```r
library(ggplot2)
Delta_Sqr_df <- as.data.frame(Delta_Sqr)
ggplot(Delta_Sqr_df, aes(x = Delta_Sqr)) + geom_density()
```

The density plot of the values looks very similar to a $\chi_2^2$ distribution. This is not surprising since the mahalanobis distance is normalizing every entry to have the same sample point to have the same variance.

## d)

```r
set.seed(1992)
library(MASS)

C <- rchisq(n=n, df = 2)
den3d <- kde2d(Delta_Sqr, C)
persp(den3d, box=FALSE)
```

We have a 3d density plot with the theoretical distribution that looks very similar to what we'd expect for $\chi_2^2$ distributions.
